```python
# estimate the gradient
def computeGradient(f, x, h):
    n = len(x)
    gradient = np.zeros(n)

    for i in range(n):
        xUp = x.copy() #make shallow copy of point X... point to values of vector x hence it's mutable... makes new copy everytime we iterate
        xUp[i] += h #add change in h to the point we are viewing in our shallow copy
        gradient[i] = (f(xUp) - f(x))/h #calculate gradient for current point by using the previous line of code
    return gradient #give back the slope of our point then repeat until calculate for every point in vector x

def gradientDescent (f, x0, alpha, h , tolerance, maxIterations): # run gradient descent and output the coordinates of the estimated critical poin
    x = x0  #set x equal to the initial guess
   
    for i in range(maxIterations):
        gradient = computeGradient(f, x, h) #calls function we just made to get the change of all relevant slopes
        if np.linalg.norm(gradient) < tolerance: #normalize (look up) gradients  and check if less than predetermined tolerance
            print('Gradient descent took' , i, 'iterations to converge')
            print('The norm of the gradient is', np.linalg.norm(gradient))
            return x #return the approximate critical x point minima
        elif i == maxIterations - 1:
            print("Gradient descent failed")
            print("The gradient is", gradient)
            return x
        x -= alpha*gradient



```

```python
# test x^2
f = lambda x : x[0]**2 #making a function of x^2 for every point we use in our vector
x = gradientDescent(f,[2],0.4,0.4,0.000001,10000) #example feeding info into cell 2
print(x, f(x))
```

```python
import os
import pandas as pd

# Function to manually map folder names to indices
def map_folder_to_index(folder_name):
    folder_mapping = {
        '70ETOH': 0,
        'AML': 1,
        'BSA': 2,
        'CPW': 3,
        'ETOH': 4,
        'Glucose': 5,
        'H20': 6,
        'HB': 7,
        'HSF': 8,
        'ISO': 9,
        'PBS': 10,
        'Syn6Sal': 11,
        'SynSal': 12,
        'Tube': 13,
        'ZMTX': 14
    }
    return folder_mapping.get(folder_name, None)

# Function to process a single file and extract the desired column
def process_file(file_path, folder_index):
    df = pd.read_csv(file_path, header=None, delimiter='\t')
    extracted_column = df.iloc[:, 2].values  # Extracting the third column
    data_to_append = [folder_index] + extracted_column.tolist()
    print(f"Data Appended: {data_to_append}")
    return data_to_append

# Function to process all files in a folder
def process_folder(folder_path):
    folder_name = os.path.basename(folder_path)
    folder_index = map_folder_to_index(folder_name)
    if folder_index is not None:
        folder_data = []
        print(f"Processing Folder: {folder_name} (Index: {folder_index})")
        for file_name in os.listdir(folder_path):
            if file_name.endswith(".txt"):
                file_path = os.path.join(folder_path, file_name)
                print(f"Processing File: {file_name}")
                folder_data.append(process_file(file_path, folder_index))
        return folder_data
    else:
        return []

# Main function to iterate through all folders
def main():
    master_data = []

    root_folder_path = r"C:\Users\Usselman Lab\VNA_model"  # Replace with the actual root folder path
    for folder_name in os.listdir(root_folder_path):
        folder_path = os.path.join(root_folder_path, folder_name)
        if os.path.isdir(folder_path):
            folder_data = process_folder(folder_path)
            master_data.extend(folder_data)

    # Create a DataFrame from the master data with transposed rows and columns
columns = ['Folder Index'] + [f'Data Point {i}' for i in range(16000)]
df_master = pd.DataFrame(master_data, columns=columns).T

# Save the transposed DataFrame to an Excel file
output_excel_path = r"C:\Users\Usselman Lab\VNA_Practice\MasterSheet.xlsx"  # Replace with the desired output path
df_master.to_excel(output_excel_path, index=False, header=False)

if __name__ == "__main__":
    try:
        main()
    except Exception as e:
        print(f"An error occurred: {e}")
```
